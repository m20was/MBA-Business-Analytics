```python
#Q.1
# Prime Number between two numbers
# Create a function prime_factors
# prime_factors(2,19) = [2, 3, 5, 7, 11, 13, 17, 19]
```

```python
def prime_factors(x,y):
    num = []
    for i in range(x,y+1):
        div = []
        for j in range(1,i+1):
            if i%j == 0:
                div.append(i)      
        if len(div) == 2:
            num.append(i)
        else:
            pass
    print(num)
      
prime_factors(2,45) 
```

```python
#Q.3
#Calculate Factorial of a number
# factorial 5 : 5*4*3*2*1  = 120
# Recursive function
```

```python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

```python
#Q.4
# Python program to check if year is a leap year or not
# Define a function is_leap_year and call it
# Pass the input in an array and call the function with a for loop
```

```python
def is_leap_year(y):
    if (y % 4) == 0:
        if (y % 100) == 0:
            if (y % 400) == 0:
                print('leap')
            else:
                print('no leap')
        else:
            print('leap')
    else:
        print('no leap')

year = [2000, 1998, 1678, 2019, 2020]

for i in range(len(year)):
    is_leap_year(year[i])
```

```python
#Q.5
# Compute HCF
# compute_hcf(320, 132)
```

```python
def compute_hcf(a, b):
    if(b == 0):
        return abs(a)
    else:
        return compute_hcf(b, a % b)
        
print(compute_hcf(320, 132))
```
